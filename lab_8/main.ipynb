```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
spotify = pd.read_csv('spotify_songs_dataset.csv')
```

```python
spotify = spotify[:50]
```

```python
spotify.head()
```

```python
spotify.info()
```

```python
plt.grid()
plt.plot(spotify['song'], spotify['danceability'])
plt.plot(spotify['song'], spotify['energy'], label='Energy', marker='s')
plt.xlabel('Danceability')
plt.ylabel('Energy')
plt.title('Danceability vs Energy (First 50 Data Points)')
plt.xticks(rotation=90, ha='right')
plt.show()
```

```python
plt.grid()
plt.scatter(spotify['danceability'], spotify['energy'])

plt.xlabel("Danceability")
plt.ylabel("Energy")
plt.title("Scatter Plot of Danceability vs. Energy")
plt.grid(True)

plt.show()
```

```python
plt.hist(spotify['popularity'])

plt.xlabel("Popularity")
plt.ylabel("Frequency")
plt.title("Histogram of Song Popularity")
plt.grid(axis='y', linestyle='--', alpha=0.7)

plt.show()
```
